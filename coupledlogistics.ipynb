```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
## defining logistic function
## r = mu = rate
## x(n+1) = μ*x(n)*(1 − x(n))

def logistic(r, x):
    return r * x * (1 - x)

## l = lambda

def mu1(l,y):
    return l*(3*y+1)

def mu2(l,y):
    return l*(-3*y+4)
```

```python
## SYMBIOSIS ##

# x0=initial pop. species 1
x0=0.5

# y0=initial pop. species 2
y0=0.5

# l=mutual benefit=lambda
l=1.393

# generations
n=30

# funtion for ploting population vs generation

x=np.zeros(n+1)
y=np.zeros(n+1)

def pop_gen_plot(x0, y0, l, n):
    
    x[0]=x0
    y[0]=y0
    
    for i in range(n):
        temp=x[i]
        x[i+1]=logistic(mu1(l,y[i]),x[i])
        y[i+1]=logistic(mu1(l,temp),y[i])
      
        
## Function calling

pop_gen_plot(x0, y0, l, n)      
  
fig, (ax1,ax2)=plt.subplots(1, 2, figsize=(12,6), sharey=True)

ax1.plot(x, 'k', lw=3, color='red')
ax1.set_title("x-Population vs Generation")
ax1.set_xlabel('Generation (n)')
ax1.set_ylabel('Population')
ax1.grid()


ax2.plot(y, 'k', lw=3, color='blue')
ax2.set_title("y-Population vs Generation")
ax2.set_xlabel('Generation (n)')
ax2.grid()


plt.tight_layout()

```

```python
## SYMBIOSIS ##

# x0=initial pop. species 1
x0=0.2

# y0=initial pop. species 2
y0=0.2

# generations
n=500

# setting step-size
div=200

# array of different values of l
l_range=np.linspace(0, 1.0843, div)

# difining arrays
R=[]
X=[]
Y=[]

# funct for calculating different values x-species & y-species at different l

def bifurcation_plot(x0, y0, l_range, n):
    

    for l in l_range:
        x=x0
        y=y0
        for i in range(n):
            if i>=250:
                R.append(l)
                X.append(x)
                Y.append(y)
                temp=x
                x=logistic(mu1(l,y),x)
                y=logistic(mu1(l,temp),y)

    
## Function calling

bifurcation_plot(x0, y0, l_range, n)      

## funtion for ploting population vs generation  
fig, (ax1,ax2)=plt.subplots(1, 2, figsize=(12,6), sharey=True)

ax1.plot(R, X, 'k', lw=1, color='red')
ax1.set_title("x-Population vs l")
ax1.set_xlabel('Mutual Benefit (l)')
ax1.set_ylabel('Population')
ax1.grid()
ax1.set_ylim(0,1)

ax2.plot(R, Y, 'k', lw=1, color='blue')
ax2.set_title("y-Population vs l")
ax2.set_xlabel('Mutual Benefit (l)')
ax2.grid()


plt.tight_layout()
```
